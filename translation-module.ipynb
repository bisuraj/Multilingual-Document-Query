```python
from easygoogletranslate import EasyGoogleTranslate
import pdfx
from reportlab.lib.pagesizes import letter
from reportlab.platypus import SimpleDocTemplate, Paragraph
from reportlab.platypus import Spacer
from reportlab.lib.styles import getSampleStyleSheet

# Load PDF and translate text
pdf = pdfx.PDFx("/content/drive/MyDrive/Colab Notebooks/tamil.pdf")
translator = EasyGoogleTranslate(
    source_language='ta',
    target_language='en',
    timeout=10
)
text = pdf.get_text()
translated_text = translator.translate(text)

# Create a PDF file
output_pdf = "/content/drive/MyDrive/Colab Notebooks/translated_text.pdf"

# Split translated text into paragraphs
translated_text = translated_text.replace('\n', ' ').strip()
paragraphs = translated_text.split('.')

# Create a PDF with translated text
doc = SimpleDocTemplate(output_pdf, pagesize=letter)
story = []

styles = getSampleStyleSheet()
style = styles["Normal"]

for paragraph in paragraphs:
    p = Paragraph(paragraph, style)
    story.append(p)
    story.append(Spacer(1, 12))

doc.build(story)

print(f"Translated text saved to: {output_pdf}")
```
